```python
class Student :
    def __init__(self,rollno,name) :
        self.roll_no = rollno
        self.name = name

    def __str__(self):
        print("Now it is in the object ")
        # here is the change
        zipped = zip(self.roll_no,self.name)
        print(zipped) # we can't use 'int' as integers are not iterable , but can we create their enumerate()
        return "nothing"


if __name__ == '__main__' :
    s1 = Student(1,"AT")
    print(s1)
```
