# Customize custom exceptions

```python
class Able2vote(Exception):
    pass


def Able2vote(age):
    if age >= 18:
        print("You are eligible to vote")
    else :
        print("You are not eligible to vote")
        raise Able2vote

Able2vote(8)
```

```python
class StudentNotFound(Exception):
    def __init__(self,roll_no):
        self.roll_no = roll_no
        super().__init__("Student not found")

def searchStudent(roll_no):
    if roll_no < 0 or roll_no > 100:
        raise StudentNotFound(roll_no)
    else :
        print(" Student found ")

searchStudent(8)
searchStudent(800)
```
