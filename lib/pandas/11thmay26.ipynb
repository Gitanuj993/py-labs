dict and datafram

```python
import pandas as pd
my_dict = {'Name' : ["Anuj","Prem","Shivkumar","Neha","Yash"],
           'age': [19,20,19,20,20]}
df = pd.DataFrame(my_dict)
print(df)
```

dict , dataframes with labelled index

```python
import pandas as pd
my_dict = {'Name' : ["Anuj","Prem","Shivkumar","Neha","Yash"],
           'age': [19,20,19,20,20]}
df = pd.DataFrame(my_dict,index=['a','b','c','d','e',])
print(df)
```
