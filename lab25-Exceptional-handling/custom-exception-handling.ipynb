# Customize custom exceptions

```python
class Able2vote(Exception):
    pass


def Able2vote(age):
    if age >= 18:
        print("You are eligible to vote")
    else :
        print("You are not eligible to vote")
        raise Able2vote

Able2vote(8)
```

```python
class StudentNotFound(Exception):
    def __init__(self,roll_no):
        self.roll_no = roll_no
        super().__init__("Student not found")

def searchStudent(roll_no):
    if roll_no < 0 or roll_no > 100:
        raise StudentNotFound(roll_no)
    else :
        print(" Student found ")

searchStudent(8)
searchStudent(800)
```

```python
class UnsufficientBalanceError(Exception):
    def __init__(self,balance):
        self.bal= balance
        super().__init__(f"UnsufficientBalance, balance : {balance} !")

class NegativeBalanceError(Exception):
    def __init__(self):
        super().__init__(" Amount can not be negative")

bal = 20000 # why bal which is a global variable can't access , use global bal
def withdraw(bal,amount):
    if amount > bal :
        raise UnsufficientBalanceError(bal)
    elif amount <= 0 :
        raise NegativeBalanceError
    else :
        bal -= amount
        print(f" Amount : {amount} debited successfully ! ")
        print(f" balance left is : {bal}")


# Input validation
if __name__ == "__main__" :
    try :
        amount = int(input("Enter amount to withdraw : "))
    except ValueError :
        print(" Please Enter numbers only !")
    else : # else can't be with try block , else came in last
        withdraw(bal,amount)
    finally:
        print(" Resources released successfully !,")

```
