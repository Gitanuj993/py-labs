boolean indexing

```python
import numpy as np

arr = np.array([10.20,30,40,50,60])
boolean_index_1 = arr[[True,True,True,False,False]]
print(boolean_index_1)
boolean_index_2 = arr[arr > 10]
print(boolean_index_2)
boolean_index_3 = arr[arr < 10]
print(boolean_index_3)

boolean_index_4  = arr[True]
print(boolean_index_4)

boolean_index_5  = arr[False]
print(boolean_index_5)

```

np.where( <condition> )

```python
import numpy as np

arr = np.array([
    [1,2,3],
    [5,6,7],
    [8,9,0]
])
result = np.where(arr > 5)
print(result)
```

condition on 2D array , filtering.

```python
import numpy as np

arr = np.array([1,2,3])
result = np.where(arr > 5)
print(result)
result = np.where(arr > 2)
print(result)
```

```python
import numpy as np

a = np.array([1, 2, 3, 4])

print(np.add(a, 5))       # [6 7 8 9]
print(np.subtract(a, 1))  # [0 1 2 3]
print(np.multiply(a, 2))  # [2 4 6 8]
print(np.divide(a, 2))    # [0.5 1. 1.5 2.]
```
